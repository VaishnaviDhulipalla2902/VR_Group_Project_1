```python
def load_cfar10_train(path, batch_id):
    with open(path + '/data_batch_' + str(batch_id), mode='rb') as file:
        batch = pickle.load(file, encoding='latin1')
        
    features = batch['data'].reshape((len(batch['data']), 3, 32, 32)).transpose(0, 2, 3, 1)
    labels = batch['labels']
        
    return features, labels

def load_cfar10_test(path):
    with open(path + '/test_batch', mode='rb') as file:
        batch = pickle.load(file, encoding='latin1')

    features = batch['data'].reshape((len(batch['data']), 3, 32, 32)).transpose(0, 2, 3, 1)
    labels = batch['labels']
        
    return features, labels
```

```python
print("Loading Training Dataset.")
images_train = []
labels_train = []
#features_part_train, labels_part_train = load_cfar10_train("data2b", 1)
for i in range(1,6):
    features_part_train, labels_part_train = load_cfar10_train("CIFAR10", i)
    images_train.extend(features_part_train)
    labels_train.extend(labels_part_train)
    print("Loaded batch", i)

images_train = np.array(images_train)
labels_train = np.array(labels_train)

#images_train = np.array(features_part_train)
#labels_train = np.array(labels_part_train)

print("Loading Test Dataset.")
images_test, labels_test = load_cfar10_test("CIFAR10")
print("Loaded Test Dataset.")
```

```python
def cvtToGray(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    resized_image = cv2.resize(gray, (150, 150), interpolation=cv2.INTER_AREA) 
    return resized_image
```

```python
def extractSIFTIndividual(image):
    sift = cv2.xfeatures2d.SIFT_create()
    _, desc = sift.detectAndCompute(image,None)
    return desc
```

```python
def extractSIFTFeatures(image_list):
    length = len(image_list)
    count = 0

    featuresBeforeClustering = []  
    for image in image_list:
        count += 1
        progress_percent = (count/length)*100
        print ("count =", count)
        print ("Length =", length)
        print('SIFT Extraction Progress: %6.2f%%'%(progress_percent),end='\r')
        
        image = cvtToGray(image)
        desc = extractSIFTIndividual(image)

        if desc is not None:
           # if featuresBeforeClustering is None:
            featuresBeforeClustering.append(desc)
           # else:
           #    featuresBeforeClustering = np.append(featuresBeforeClustering,desc,axis=0)
    featuresBeforeClustering = np.concatenate(featuresBeforeClustering, axis=0)
#     featuresBeforeClustering = np.asarray(featuresBeforeClustering)
    print ('SIFT extraction done!')
    
    return featuresBeforeClustering
```

```python
def kMeans(features):
    print("KMeans clustering begins.")
    
    result = MiniBatchKMeans(init='k-means++', n_clusters=48,max_iter=1000, batch_size=1000, n_init=10, max_no_improvement=10, verbose=0).fit(features)

#     term_criteria = (cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

#     ret, label, center = cv2.kmeans(features, 16, None, term_criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    print("KMeans clustering ends.")

    return result

# labels_kmeans, centers_kmeans = kMeans(dataset)
# labels_kmeans_test, centers_kmeans_test = kMeans(dataset_test)
# print("done")
```

```python
def genVLADIndividual(image, k_means_model):
    
    predictedLabels = k_means_model.predict(image)
    clusterCenters = k_means_model.cluster_centers_
    lebels = k_means_model.labels_
    numClusters = k_means_model.n_clusters
    
    r,d = image.shape
    result = np.zeros([numClusters,d])
    
    for i in range(numClusters):
        progress_percent = (i/numClusters)*100
        print('Individual VLAD Generation Progress: %6.2f%%'%(progress_percent),end='\r')
        
        if (np.sum(predictedLabels == i) > 0):
            result[i] = np.sum(image[predictedLabels == i, :] - clusterCenters[i], axis=0)
    result = result.flatten()
    
    result = np.sign(result)*np.sqrt(np.abs(result))

    # L2 normalization

    result = result/np.sqrt(np.dot(result,result))
    return result
    
```

```python
def genVLADFeatures(images, model):
    descriptors = []
    length = len(images)
    
    count = 0
    for image in images:
        count += 1
        progress_percent = (count/length)*100
        print('VLAD Generation Progress: %6.2f%%'%(progress_percent),end='\r')
        
        image = cvtToGray(image)
        des = extractSIFTIndividual(image)
        
        if des is not None:
            vDes = genVLADIndividual(image, model)
            descriptors.append(vDes)
            
    descriptors = np.asarray(descriptors)
    return descriptors
            
```

```python
features = extractSIFTFeatures(images_train)
model_mini_kmeans = kMeans(features)
```
